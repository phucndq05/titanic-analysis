```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice style for the plots
sns.set_style("whitegrid")

# Load the dataset from the 'data' folder
df = pd.read_csv('data/train.csv')

# Display the first 5 rows to check if the data loaded correctly
df.head()
```

```python
# Question 1: How did gender affect survival rate?
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Status', labels=['Did not survive', 'Survived'])
plt.show()
```

```python
# Question 2: How did passenger class affect survival rate?
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Status', labels=['Did not survive', 'Survived'])
plt.show()
```

```python
# --- CELL 4: Deeper Analysis on Age and Survival ---

# 1. Histogram to see the age distribution of survivors vs. non-survivors
g = sns.FacetGrid(df, col='Survived', height=5)
g.map(sns.histplot, 'Age', bins=20, kde=False)
plt.suptitle('Age Distribution by Survival Status', y=1.02)
plt.show()


# 2. Violin plot to see the age distribution combined with gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df, split=True, palette='pastel')
plt.title('Age Distribution by Sex and Survival', fontsize=16)
plt.show()
```
